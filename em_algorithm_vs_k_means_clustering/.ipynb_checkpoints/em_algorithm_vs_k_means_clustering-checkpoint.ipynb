```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

import pdb
```

```python
df = pd.read_csv("dataset/data.csv")

distance_feature = df['Distance_Feature'].values
speeding_feature = df['Speeding_Feature'].values
```

```python
X = np.matrix(list(zip(distance_feature,speeding_feature)))
```

```python
figure = plt.figure()
d,k,g = figure.subplots(3,1)
d.set_xlim([0,100])
d.set_ylim([0,50])
d.set_title('Visualizing dataset')
d.set_ylabel('Speeding Feature')
d.set_xlabel('Distance Feature')
d.scatter(distance_feature,speeding_feature)
```

```python
colors = ['r','g','b']
markers = ['o','s','v']
plt.plot(2)
plt.subplot(513)
kmeans_model = KMeans(n_clusters = 3).fit(X)

for i, j in enumerate(kmeans_model.labels_):
    plt.plot(distance_feature[i],speeding_feature[i],color=colors[j],marker=markers[j])

plt.xlim([0,100])
plt.ylim([0,50])
plt.title('Visualizing K-means clustering')
plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')

plt.plot(3)
plt.subplot(515)
gmm = GaussianMixture(n_components = 3).fit(X)
labels = gmm.predict(X)

for i, j in enumerate(labels):
    plt.plot(distance_feature[i],speeding_feature[i],color=colors[j],marker=markers[j])

plt.xlim([0,100])
plt.ylim([0,50])
plt.title('Visualizing GaussianMixture clustering')
plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')
```

```python
plt.show()
```
