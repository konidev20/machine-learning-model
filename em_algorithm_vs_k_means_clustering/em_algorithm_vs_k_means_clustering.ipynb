```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

import pdb
```

```python
df = pd.read_csv("dataset/data.csv")

distance_feature = df['Distance_Feature'].values
speeding_feature = df['Speeding_Feature'].values
```

```python
X = np.matrix(list(zip(distance_feature,speeding_feature)))
```

```python
figure1 = plt.figure()
d = figure1.subplots(1,1)
d.set_xlim([0,100])
d.set_ylim([0,50])
d.set_title('Visualizing dataset')
d.set_ylabel('Speeding Feature')
d.set_xlabel('Distance Feature')
d.scatter(distance_feature,speeding_feature)
```

```python
colors = ['r','g','b']
markers = ['o','s','v']
```

```python
figure2 = plt.figure()
k = figure2.subplots(1,1)
kmeans_model = KMeans(n_clusters = 3).fit(X)
labels = kmeans_model.predict(X)

for i, j in enumerate(labels):
    k.plot(distance_feature[i],speeding_feature[i],color=colors[j],marker=markers[j])

k.set_xlim([0,100])
k.set_ylim([0,50])
k.set_title('Visualizing K-means clustering')
k.set_ylabel('Speeding Feature')
k.set_xlabel('Distance Feature')
```

```python
figure3 = plt.figure()
g = figure3.subplots(1,1)
gmm = GaussianMixture(n_components = 3).fit(X)
labels = gmm.predict(X)

for i, j in enumerate(labels):
    g.plot(distance_feature[i],speeding_feature[i],color=colors[j],marker=markers[j])

g.set_xlim([0,100])
g.set_ylim([0,50])
g.set_title('Visualizing GaussianMixture clustering')
g.set_ylabel('Speeding Feature')
g.set_xlabel('Distance Feature')
```
