```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def kernel(point, X, k):
    m,n = np.shape(X)
    w = np.mat(np.eye(m))
    for j in range(m):
        d = point - X[j]
        w[j,j] = np.exp(d * d.T / (-2.0*k**2))
    return w
```

```python
def localWeight(point,X,Y,k):
    w = kernel(point,X,k)
    W = (X.T*(w*X)).I * (X.T*(w * Y.T))
    return W
```

```python
def locallyWeightedRegression(X,Y,k):
    row,col = np.shape(X)
    ypred = np.zeros(row)
    for i in range(row):
        ypred[i] = X[i] * localWeight(X[i],X,Y,k)
    return ypred
```

```python
data = pd.read_csv("data.csv")

bill = np.array(data.total_bill)
tip = np.array(data.tip)
```

```python
bill_matrix = np.mat(bill)
tip_matrix = np.mat(tip)
```

```python
row,col = np.shape(bill_matrix)
ones_array = np.mat(np.ones(col))
X = np.hstack((ones_array.T,bill_matrix.T))
```

```python
ypred = locallyWeightedRegression(X,tip_matrix,2)
```

```python
fig= plt.figure()
ax=fig.subplots(1,1)
ax.scatter(bill,tip,color='blue')
plt.xlabel('Total bill')
plt.ylabel('tip')

s = X[:,1].argsort(0)
xsort = X[s][:,0]

ax.plot(xsort[:,1],ypred[s],color="red",linewidth=1)
```
